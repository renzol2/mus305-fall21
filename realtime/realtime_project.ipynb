```python
import threading, time, sys
import rtmidi
import musx
print(musx.version)
```

```python
midiout = rtmidi.MidiOut()
print(midiout)
outports = midiout.get_ports()
print("available ports:", outports) 
```

```python
loop_midi_outport = 'loopMIDI Port 3'  # find loop midi
midiout.open_port(outports.index(loop_midi_outport))
```

```python
midiin = rtmidi.MidiIn()
print(midiin)
inports = midiin.get_ports()
print("available input ports:", inports)
```

```python
midi_controller_port = 'Mobile Keys 49 0'  # find midi controller
midiin.open_port(inports.index(midi_controller_port))
```

```python
from musx import Pitch, Interval
def arpeggiator(message, data):
    print(message)
    NOTE_INDEX = 1
    NOTE_ON = 144
    root = message[0][NOTE_INDEX]
    is_note_on = message[0][0] == NOTE_ON

    maj_chord_intervals = [Interval('M3'), Interval('P5'), Interval('M7')]
    root_pitch = Pitch.from_keynum(root)
    maj_chord = [root_pitch] + [i.transpose(root_pitch) for i in maj_chord_intervals]

    print(maj_chord)
    
    if is_note_on:
      for pitch in maj_chord:
        midiout.send_message(musx.note_on(0, pitch.keynum(), 80))
        time.sleep(0.1)
        midiout.send_message(musx.note_off(0, pitch.keynum(), 127))
    else:
      for pitch in maj_chord:
        midiout.send_message(musx.note_off(0, pitch.keynum(), 127))

    # time.sleep(.5)
  
midiin.set_callback(arpeggiator)
```
